```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text

np.random.seed(42)

rows=[]
for _ in range(300):
    prev=np.random.uniform(25,40)
    now=prev+np.random.uniform(-0.2,1.2)
    rise=now-prev
    time=np.random.uniform(0,200)

    if now>37: label=1
    elif rise>0.7: label=1
    elif time>120: label=1
    else: label=0

    rows.append([now,rise,time,label])

df=pd.DataFrame(rows,columns=["temp_now","temp_rise","motor_time","label"])

X=df[["temp_now","temp_rise","motor_time"]]
y=df["label"]

tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X,y)

print(export_text(tree, feature_names=list(X.columns)))
```
